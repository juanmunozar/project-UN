```python
import matplotlib.pyplot as plt
import numpy as np

# Filter the dataframe for 'ENFERMERÍA' program
engineering_df = df2[df2['PROGRAMA ACADÉMICO'] == 'ENFERMERÍA']

grouped_df = engineering_df.groupby(['DEPARTAMENTO DE OFERTA DEL PROGRAMA', 'SEXO'])['INSCRITOS 2016'].sum()

# Reshape the data to create a matrix for the heatmap
heatmap_data = grouped_df.unstack(level='SEXO')
heatmap_data = heatmap_data.astype(int)

# Create a heatmap
plt.figure(figsize=(10, 8))
plt.imshow(heatmap_data, cmap='YlGnBu', aspect='auto')

# Set ticks and tick labels
plt.xticks(np.arange(2), ['Female', 'Male'])
plt.yticks(np.arange(len(heatmap_data)), heatmap_data.index)

# Add colorbar
cbar = plt.colorbar()
cbar.set_label('Total Enrollments')

# Add text annotations
for i in range(len(heatmap_data)):
    for j in range(2):
        plt.text(j, i, str(heatmap_data.iloc[i, j]), ha='center', va='center', color='black')

# Set plot title and axis labels
plt.title('Enrollment in ENFERMERÍA Program by Department and Gender')
plt.xlabel('Gender')
plt.ylabel('Department')
plt.savefig('heatmap.png')

# Display the plot
plt.show()
```
